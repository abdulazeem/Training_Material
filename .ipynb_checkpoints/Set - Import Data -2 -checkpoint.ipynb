# Exercise 5: Parsing HTML with BeautifulSoup


1) Import https://www.crummy.com/software/BeautifulSoup/bs4/doc/ and parse using BeautifulSoup

2) Prettify the raw data using .prettify() method

3) Find the structure of webpage.

4) Extract title of the document from the data using .title() method

5) Extract text from the data using get_text() method

6) Using .find and .findall on the html tags

7) Use .select to extract the tags


```python
from bs4 import BeautifulSoup
import requests

url= 'https://en.wikipedia.org/wiki/List_of_deepest_caves'
r = requests.get(url)

data=r.text

soup = BeautifulSoup(data,'lxml')

#Prettify the input data
print(soup.prettify())
```

```python
#Find the structure of webpage

struct = [type(item) for item in soup.children]
print(struct)
```

```python
#Find the title of the webpage

title = soup.title
title_text=soup.title.text
print(title)
print(title_text)
```

```python
struct = [item for item in soup.children]
struct[1].get_text()
```

```python
#Using .find and .findall on the html tags

a_tags = soup.find_all('a')

p_tags = soup.find_all('p')

div= soup.find_all('div', class_='noprint')

table = soup.find_all('table')

print(table)
```

```python
#Using select on the html tags

a_tags = soup.select('a')

p_tags = soup.select('p')

div= soup.select('div')

table = soup.select('table')

print(table)
```

# Exercise 6: Parsing HTML with BeautifulSoup - Getting all hyperlinks using tags

Scrape the website to extract all Hyperlinks

https://en.wikipedia.org/wiki/Isaac_Newton 

```python
import requests
from bs4 import BeautifulSoup

url = 'https://en.wikipedia.org/wiki/Isaac_Newton'
req=requests.get(url)
data=req.text
soup=BeautifulSoup(data)

a_tags= soup.find_all('a')

for link in a_tags:
    print(link.get('href'))
```

# Exercise 7: Import Table from Webpage

Import the titles of all the Asian countries from the given webpage

'https://en.wikipedia.org/wiki/List_of_Asian_countries_by_area'

```python
import requests
from bs4 import BeautifulSoup

req=requests.get('https://en.wikipedia.org/wiki/List_of_Asian_countries_by_area')
data= req.text

soup=BeautifulSoup(data)

new_table=soup.find('table', class_='wikitable sortable')
```

```python
links=new_table.find_all('a')

lst=[]
for link in links:
    lst.append(link.get('title'))
    
lst
```
