# Exercise 6: Categorical - Countplot using Catplot()

#### 1) Find the reported usage of Internet amongst the young people from the survey data.
    
    a) Use catplot with x axis containing the usage data to create a count plot.
    
    b) Use catplot with y axis containing the usage data to create a count plot.
    
    c) Create subplots using 'row' and 'column' arguments for the Gender and Village-town column.
    
    d) Create columns subplots based on 'Age' category. The categorical column should have only two categories. 'Under 21' '21 and Above'


```python
survey=pd.read_csv('young-people-survey.csv')
survey.head()
```

```python
# a) Use catplot with x axis containing the usage data
# to create a count plot.
# b) Use catplot with y axis containing the usage data
# to create a count plot.
    
import seaborn as sns
import matplotlib.pyplot as plt

sns.catplot(y='Internet usage', data=survey, kind='count')
plt.show()
```

```python
# c) Create subplots using 'row' and 'column' arguments 
# for the Gender and Village-town column.
sns.catplot(y='Internet usage', data=survey, kind='count', row='Village - town', col='Gender')
#plt.show()
```

```python
# d) Create columns subplots based on 'Age' category. 
# The categorical column should have only two categories. 
# 'Under 21' '21 and Above'


survey['Age Category']=(survey['Age']<21).map({True: 'Below 21', False: '21 and Above'})
```

```python
sns.catplot(y='Internet usage', data=survey, kind='count',  col='Age Category')
plt.show()
```

# Exercise 8: Categorical - Boxplot using Catplot()

Use student-alcohol.csv

1) Create a boxplot to view the five point summary statistics for the Grades 1, based on the study_time. What does the boxplot reveal?

2) Find the distribution of grades G3 based on the internet access available to the students. 

    a) Additionally add the distribution based on location using 'hue' argument.
    
    b) Additionally, turn the outliers, using sym=''.
    
3) Find the effect of being in 'romance' relationship on the grades G3. 

    a) Adjust the boxplot to have whiskers to extend to 0.5*IQR
    
    b) Ajust whiskers to show 5th and 95th percentiles.
    
    c) Adjust whiskers to show min and max values.
    
    Note: Default setting of boxplot is to have whiskers of size 1.5*IQR 


```python
# 1) Boxplot of G1 based on study time dedicated by students

order=['<2 hours', '2 to 5 hours', '5 to 10 hours', '>10 hours']
sns.catplot(x='study_time', y='G1', data=student, kind='box', order=order)
plt.show()
```

```python
student.columns
```

```python
# 2) Boxplot of G3 based on internet availability. 

# hue argument as location
# Switch off outliers with sys argument

sns.catplot(x='internet', y='G3', data=student, kind='box', hue='location', sym='')
```

```python
# 3) Boxplot : Effect of relationships on results

sns.catplot(x='romantic', y='G3', data=student, kind='box')
```

```python
# 3a) Adjusting whiskers to have 0.5*IQR

sns.catplot(x='romantic', y='G3', data=student, kind='box', whis=0.5)
```

```python
# 3b) Adjusting whiskers to extend to 5th and 95th percentiles

sns.catplot(x='romantic', y='G3', data=student, kind='box', whis=[5,95]) 
```

```python
# 3c) Adjusting whiskers to show min and max values.

sns.catplot(x='romantic', y='G3', data=student, kind='box', whis=[0,100]) 
```

```python
# 3c) Adjusting whiskers to show min and max values.

sns.catplot(x='romantic', y='G3', data=student, kind='box', whis=[0,100]) 
```

```python
# 1) Set Style: Change Style of the figure using sns.set_style()
        
# 2) Set Palette: Change palette using sns.palette()

sns.set_style('whitegrid')

sns.set_palette('Blues')
sns.countplot(y='Team', data=bowling)
plt.show()
```

```python
# 3) Set Scale using sns.set_context: Change scale using set.set_context()


sns.set_style('darkgrid')

#sns.set_palette('RdBu')

sns.set_context('paper')
custom=['red','blue']

sns.set_palette(custom)
sns.countplot(x='Bowling Hand', data=bowling)
plt.show()
```

# Adding Title to the Plot

1) Create a scatter plot relation between the Overs and Runs in IPL Match using sns.scatterplot

    a) Identify the type of the plot.
    
    b) Add the following title, 'Overs vs Runs' usng .set_title()
    
    c) Modify the x and y label, using .set(xlabel, ylabel)
    
    d) Rotate the xticks by 90 degree using plt.xticks(rotation=90)
    
2) Using relplot create a scatter plot relation between the Overs and Runs in IPL Match.

    a) Identify the type of the plot.
    
    b) Add the following title, 'Overs vs Runs' usng .fig.suptitle()
    
    c) For the supplots use .set_titles(' Name: {col_names})


```python
# 1

a=sns.scatterplot(x='Runs', y='Wkts', data=bowling)
type(a)

a.set_title('Overs vs Runs', y=1.02)
a.set(xlabel='Runs Scored', ylabel='Wickets Taken')

plt.xticks(rotation=90)
```

```python
# 2

a=sns.catplot(y='Wkts', data=bowling, kind='box', sym='', col='Bowling Hand')

type(a)

a.fig.suptitle('Overs vs Runs', y=1.02)

a.set_titles('Hand: {col_name}')


a.set(xlabel='Hand', ylabel='Wickets')
```
