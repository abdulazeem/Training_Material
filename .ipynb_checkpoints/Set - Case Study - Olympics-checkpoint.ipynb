```python
import pandas as pd
olympics= pd.read_csv('olympics.csv')
olympics.head()
```

```python
country_medals = olympics['NOC'].value_counts()
country_medals[:10]
```

```python
counted = olympics.pivot_table(index='NOC', values='Athlete', columns='Medal', aggfunc='count')
counted['total']=counted.sum(axis=1)
counted.head(20)
```

```python
counted.sort_values('total', ascending=False)
```

```python
count_sport=olympics.groupby(['NOC'])[['Sport']].nunique()
count_sport.sort_values('Sport', ascending=False)
```

```python
count_sport=olympics.groupby(['Sport'])[[ 'NOC']].nunique()
count_sport.sort_values('NOC', ascending=False)
```

```python
olympics.groupby(['NOC', 'Medal'])[['NOC']].count()
```
