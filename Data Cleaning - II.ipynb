# Exercise 1: Concatenation
    
1) Create two simple data frames for concatenation

4) Combine data column wise usind pd.concat() with axis=1 

5) Rename the column names of the combined data using .columns method

```python
# Create two simple data frames for concatenation

import pandas as pd

dict1={'A':[1,2,3,4], 'B':[5,6,7,8]}
df1 = pd.DataFrame(dict1)
print('Header of First Data Frame \n \n', df1.head())
print()

dict2={'A':[5,6,7,8], 'B':[1,2,3,4]}
df2 = pd.DataFrame(dict2)
print('Header of Second Data Frame \n \n', df2.head())
```

```python
# Concatenating data row wise
pd.concat([df1, df2])
pd.concat([df1, df2], ignore_index=True)
```

```python
# Concatenating data column wise 

new_df= pd.concat([df1, df2], axis=1)
new_df.columns=['A', 'B', 'C', 'D']

new_df.head()
```

# Exercise 6: Using Regular Expressions 

When working with data, it is necessary to write a regular expression to look for properly entered values. This will validate the data to be in the format which we intend. 

# re.match(pattern, data)

Use .span(), string, and .group() methods along with match to observe the start-end positions, string part passed to function, part of the string where there was a match respectively.

1) Match the word 'John1' by passing exactly same word in the pattern. Check also by writing different numbers infront of John.

2) Match a pattern to validate the total amount to be in the form $123.45 (sample amount up to two decimals only)

3) Match a pattern to validate the entry of phone number in a country to be of the form 123-456-7861 (sample number)

4) Match a pattern to validate password which should contain atleast one capital letter, one special character, one digit, with a 6-12 character length password

# re.findall(pattern, data)

5) Extract number from the string 'There are 20 different types of items in the workshop'

6) Extract numbers from the string 'The height and weight of the individual are 156.23 cm and 38.5 kg'

https://www.w3schools.com/python/python_regex.asp


```python
# Match the word John

import re

pattern = '^John\d+$'

data = 'John2'

result = re.match(pattern, data)
bool(result)
```

```python
# Match a pattern to validate the total amount to be in the form $123.45 

import re

pattern = '^\$\d*\.\d{2}$'

data = '$123.45'

result = re.match(pattern, data)
bool(result)

result.group()
```

```python
# Match a pattern to validate the entry of phone number in a country to be of the form 123-456-7861 (sample number)

import re

pattern = '^\d{3}-\d{3}-\d{4}$'

data = '123-541-2145'

result = re.match(pattern, data)
bool(result)
```

```python
# Match a pattern to validate password which should contain 
# atleast one capital letter, one special character, one digit, with a 6-12 character length password
import re

pattern = "^(?=.*[a-z])(?=.*[A-Z])(?=.*\d)(?=.*[@$!%*#?&])[A-Za-z\d@$!#%*?&]{6,20}$"

data = 'Ab$2adsfd'

result = re.match(pattern, data)
bool(result)
```

```python
# Extract number from the string 'There are 20 different types of items in the workshop'

pattern = '\d+'
data = 'There aree 20 different types of items in the workshop'

re.findall(pattern, data)
```

```python
# Extract numbers from the string 'The height and weight of the individual are 156.23 cm and 38.5 kg'

pattern = '\d+\.\d+'
data = 'The height and weight of the individual are 156.23 cm and 38.5 kg'

re.findall(pattern, data)
```

1) Check the missing values using .info() and .isnull() methods.

3) Find the percentage of missing data

4) Use fillna with statistical values to fill the missing data.

    a) Categorical data: .fillna(method='ffill')
    
    b) Numerical data : .fillna(np.mean(data))
    
5) Use interpolate(method='linear', method_direction='forward') for numerical data

Use billing_types.csv file 

```python
import pandas as pd

data = pd.read_csv('billing_types.csv')
data.info()
```

```python
data['gender']=data.gender.astype('category')
data['smoker']=data.smoker.astype('category')
data['total_bill']=pd.to_numeric(data['total_bill'], errors='coerce')
data['tip']=pd.to_numeric(data['total_bill'], errors='coerce')
```

```python
# 2) Drop duplicate data from the dataset using .dropduplicates() method and re-check the shape of the dataset.

data.info()

print(data.isnull().sum())

print(data.isnull().sum().sum())
```

```python
# Dropping Duplicates
data.drop_duplicates(inplace=True)
data.shape
```

```python
data.gender.interpolate(method='ffill', method_direction='forward')
```

```python
# 3) Find the percentage of missing data
import numpy as np

num_of_missing= data.isnull().sum().sum()
total_cells=np.product(data.shape)

Percentage= (num_of_missing/total_cells)*100
print(Percentage)
```

```python
data.shape
```

```python
# 4) Use fillna with statistical values to fill the missing data.

data.total_bill.fillna(np.mean(data.total_bill), inplace=True)

data.tip.fillna(np.mean(data.tip), inplace=True)

data.smoker.fillna(method='ffill', inplace=True)

data.gender.fillna(method='bfill', inplace=True)

data.head()
```

```python
# 5) Use interpolate(method='linear', method_direction='forward') for numerical data

data.tip.interpolate(method='linear', inplace=True)
data.head()
```

```python
# 6) Fill the missing values in bulk for categorical and numerical columns

data.fillna(method='ffill')
```

```python
# 7) Use .dropna(how='any') or .dropna(how='all') and argument axis =0/1 to drop column or row

data.dropna(how='any', axis=0, inplace=True)
```
