# Exercise 3: Column Transformation
    
### 1) Usage of .apply() function
    
    a) Use daily_weather data and modify the columns ['Mean TemperatureF', 'Min TemperatureF'] to return the celsius data.

    a) The data in arizona_students has 'Handed column. Change the full names in the column with abbreviations using .map function.
    
        - Create a dictionary of abbreviations for each category type. Use this to pass into the function .map(dictionary)
    
    
#### a) use daily_weather.csv
#### b) use arizona_students.csv

```python
df =pd.read_csv('daily_weather.csv', index_col='Date')
```

```python
df.head(2)
```

```python
def to_celsius(F):
    return 5/9*(F - 32)
```

```python
# Use .apply() function to transform columns
# Convert each value of the 'Mean TemperatureF' and 'Min TemperatureF' column to celsius value.

celsius_data=df[['Mean TemperatureF', 'Min TemperatureF']].apply(to_celsius)

celsius_data.columns=['Mean TempC', 'Min TempC']

celsius_data.head()
```

```python
# Use .map() to transform categorical variables

df=pd.read_csv('arizona_students.csv')

lst_map={'Right-Handed': 'RH', 'Left-Handed': 'LH', 'Ambidextruous': 'AMB'}

df['Handed']=df.Handed.map(lst_map)

df.head()
```
